```python
#importing all the libs which are requied 

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import hvplot.pandas
```

```python
#importing all the data from the data sets

dt1 = pd.read_csv("https://tinyurl.com/ChrisCoDV/001203994/VenueDailyVisitors.csv", index_col=0)
dt2 = pd.read_csv("https://tinyurl.com/ChrisCoDV/001203994/VenueAge.csv ", index_col=0)
dt3 = pd.read_csv("https://tinyurl.com/ChrisCoDV/001203994/VenueDistance.csv", index_col=0)
dt4 = pd.read_csv("https://tinyurl.com/ChrisCoDV/001203994/VenueDuration.csv ", index_col=0)
dt5 = pd.read_csv("https://tinyurl.com/ChrisCoDV/001203994/VenueGender.csv ", index_col=0)
dt6 = pd.read_csv("https://tinyurl.com/ChrisCoDV/001203994/VenueSpend.csv", index_col=0)
```

```python
#merging age, distance, duration, gender, spend in a single data frame 

dtall = pd.DataFrame(index=dt1.columns)
dtall['Age'] = dt2.values
dtall['Distance'] = dt3.values
dtall['Duration'] = dt4.values
dtall['Female ratio %'] = dt5.values
dtall['spend']=dt6.values
print(dtall)
```

```python
#total number of vistors graph for venues without sorting 

plt.figure(figsize=(20, 10))
x_pos = np.arange(len(dt1.columns))
plt.bar(x_pos, dt1.sum(), align='center')
plt.xticks(x_pos, dt1.columns)
plt.xlabel('Venues', fontsize=20)
plt.ylabel('Visitors', fontsize=20)
plt.title('Total Visitors for Venues unsorted', fontsize=25)
plt.show()
```

```python
#sorting of data from high to low to undestand well

dt1_sorted= dt1.reindex(dt1.sum().sort_values(ascending=False).index, axis=1)
```

```python
#sorted raw data 

print(dt1_sorted)
```

```python
dt1_sorted.sum()
```

```python
#sorted graph of vistors from high to low.

plt.figure(figsize=(18,10))

x_pos = np.arange(len(dt1_sorted.columns))
plt.bar(x_pos, dt1_sorted.sum(), align='center')
plt.xticks(x_pos, dt1_sorted.columns , rotation = 90)
plt.xlabel('Venues', fontsize=20)
plt.ylabel('Visitors', fontsize=20)
plt.title('Total Visitors for Venues sorted', fontsize=25)
plt.show()
```

```python
#histrogram of the visitors of every induvidal venues

data = pd.read_csv('https://tinyurl.com/ChrisCoDV/001203994/VenueDailyVisitors.csv', index_col=0)

counter = 1
fig = plt.figure(figsize=(20, 20))
fig.suptitle('Visitors number distributions', fontsize=20, position=(0.5, 1.0))
for name in data:
    sub = fig.add_subplot(6, 8, counter)
    sub.set_title('Venue ' + name, fontsize=10)
    sub.hist(data[name], edgecolor='w')
    counter += 1
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()
```

```python
#dividing the venues into 4 different groups

categories = ['High', 'Medium', 'Low', 'Very Low']
categories_selected = [[] for i in range(len(categories))]
for name in dt1_sorted.columns:
    total_vistors = dt1_sorted[name].sum()
    if total_vistors  > 85000:
        category = 0
    elif total_vistors  > 30000:
        category = 1
    elif total_vistors  > 20000:
        category = 2
    else:
        category = 3
    categories_selected[category].append(name)
    print(name+ ' Venue'+ ' has a '+ categories[category] + ' volume')

for i in range(len(categories)):
    print(categories[i] + ': ' + str(categories_selected[i]))

for i, selected in enumerate(categories_selected):
    plt.figure(figsize=(8, 8))
    x_pos = np.arange(len(dt1_sorted[selected].columns))
    plt.bar(x_pos, dt1_sorted[selected].sum(), align='center', color="c")
    plt.xticks(x_pos, dt1_sorted[selected].columns)
    plt.xlabel('Venues', fontsize=18)
    plt.ylabel('Visitors', fontsize=18)
    plt.title(categories[i] + ' Volume of Visitors ', fontsize=20)
    plt.show()

```

```python
# showing the top 4 highiest vistors for the venue in graph

selected = ['SJU', 'RDA','SPF', 'PXI']
print(dt1[selected].head())
 
    
plt.figure(figsize=(8, 8))
x_pos = np.arange(len(dt1[selected].columns))
plt.bar(x_pos, dt1[selected].sum(), align='center',color="g")
plt.xticks(x_pos, dt1[selected].columns)
plt.xlabel('Venues', fontsize=18)
plt.ylabel('Vistors', fontsize=18)
plt.title('Top 4 venues with highiest vistors', fontsize=20)
plt.show()
```

```python
#sctter plot to identfy the corelation between the summary data

selected = ['Age' , 'Distance', 'Duration' , 'Female ratio %',  'spend']

for i, name_i in enumerate(selected):
    for j in range(i + 1, len(selected)):
        name_j = selected[j]
        plt.figure(figsize=(8, 8))
        plt.scatter(dtall[name_i], dtall[name_j])
        plt.title('Venue at ' + name_i + ' vs Venue at ' + name_j, fontsize=20)
        plt.xlabel('Venue at ' + name_i, fontsize=18)
        plt.ylabel('Venue at ' + name_j, fontsize=18)
        plt.show()

```

```python
#histrogram of the age distribution

x_min = 15
x_max = 85
bin_width = 10
n_bins = int((bin_width + x_max - x_min) / bin_width)
print(str(n_bins) + ' bins')
bins = [(x_min + x * (bin_width + x_max - x_min) / n_bins) for x in range(int(n_bins))]
# print(bins)

plt.figure(figsize=(8, 8))
plt.hist(dtall['Age'], bins=bins, edgecolor='w')
plt.title('Age distribution', fontsize=20)
plt.show()
```

```python
#Box plot for the summary data 

selected = ['Age' , 'Distance', 'Duration' , 'Female ratio %',  'spend']

plt.figure(figsize=(8, 8))
# data[selected].boxplot()
plt.boxplot(dtall[selected], labels=selected)
plt.xlabel('', fontsize=18)
plt.ylabel('', fontsize=18)
plt.title('Box plot of the summary data', fontsize=20)
plt.show()
```
